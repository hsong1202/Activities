```python
import pandas as pd
import numpy as np
```

**Read CSV**

```python
df = pd.read_csv('big_startup_secsees_dataset.csv')
df
```

**Drop Duplicates**

```python
df = df.drop_duplicates()
df
```

**Convert funding column from string to numbers for math operations**
**Average empty columns with fillna**

```python

df['funding_total_usd'] = pd.to_numeric(df['funding_total_usd'], errors='coerce')

df['funding_total_usd'].fillna(df['funding_total_usd'].mean(), inplace=True)
df
```

**Drop Rows With Missing Data**

```python
df.dropna(subset=['name', 'homepage_url', 'country_code', 'state_code', 'region', 'city', 'category_list', ], inplace=True)
df
```
