```python
import csv

newfile = open("./Python_LoopPractice.csv")
csvreader = csv.reader(newfile)

# print(type(csvreader))
# print(csvreader)

header = next(csvreader)
# print(type(header))
# print(help(next))

header.pop()

# print(csvreader)

rows = []

for row in csvreader:
    row.pop()
    rows.append(row)
#     print(row)

# print(rows)

pokedex = []
# iterate through rows list
for row in rows:
    pokedex_entry = dict()
    # nested loop to iterate through column names
    for index, col in enumerate(header):
        key = col
        value = row[index]
        pokedex_entry[key] = value
        pokedex.append(pokedex_entry)

# print(pokedex)

for row in pokedex:
    print(row)

newfile.close()

```
