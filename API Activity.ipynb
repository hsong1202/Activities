```python
import requests
from dotenv import load_dotenv
import pandas as pd
import json
import os

load_dotenv()

api_key = os.getenv('API_KEY')

# print(api_key)

response = requests.get(f'http://api.openweathermap.org/data/2.5/forecast?id=5511077&appid={api_key}')

data = response.json()

# type(data)
# len(data)
# print(data)

for key, value in data.items():
    print(key, ':', value)

# print(data['list'])
newlist = data['list']
# len(newlist)
for item in newlist:
    print(newlist)

df = pd.DataFrame(newlist)

# print(df)

df.to_csv('Renodata.csv', index =False)

print('Data exported successfully!')

```
