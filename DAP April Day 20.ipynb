```python
import pandas as pd

df = pd.read_csv("./Sqlite/data.csv")

duration_series = df["Duration"]
# print(duration_series)

for index, duration in enumerate(duration_series):
    if duration > 120:
        df.loc[index, "Duration"] = 120
print(df.to_string())

#drop values over 120

# print(df.info())

#loop through index 

for x in df.index:
    if df.loc[x, "Duration"] > 120:
        df.drop(x, inplace=True)
print(df.to_string())

# print(df.info())
```
