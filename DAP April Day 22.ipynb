```python
import requests
from bs4 import BeautifulSoup as soup
import csv

url = "http://www.values.com/inspirational-quotes"
response = requests.get(url)

results = soup(response.content, "html.parser")
print(results.prettify())

quotes = []

# <div id ="all_quotess">some text</div> tag with the id="all _quotes"
#soup.find()
#looks at the HTML page elements, finds the first Page Element that matches the given criteria

quotes_content = results.find("div", attrs = {'id':'all_quotes'})
print(quotes_content)

for tag in quotes_content.findAll("div"):
    quote = {}
    url = "https:/:" + tag.a['href']
    quote["url"] = url
    lines = tag.img['alt'].split(" #")[0]
    quote["lines"] = lines
    theme = tag.h5.a.text
    quote["theme"] = theme
    image = tag.img["src"]
    quote["image"] = image
    quotes.append(quote)

print(quotes)

#save the quotes to a CSV file

filename = "inspirational_quotes.csv"
with open(filename, "w", newline ="") as f:
    w = csv.DictWriter(f, ["theme","url","image","lines"])
    w.writeheader()
    for quote in quotes:
        w.writerow(quote)
```
